```python
answers = []
dates = []
bylines = []
urls = []

for item in sites:


   site = requests.get(item)
   html_code = site.content
   soup = BeautifulSoup(html_code)

   answer = soup.find('div', {'id': "clue_list"})
   url = item
   str_answer = str(answer)
   answer_list = str_answer.split('<br/>')


   for x in answer_list:


    date = soup.find('h1', class_ = 'page-title')
    byline = soup.find('div', {'id': "byline"})

    answers.append(x)
    dates.append(date)
    bylines.append(byline)
    urls.append(item)
```

```python
# Tossing all data into DF to be cleaned
```

```python
df = pd.DataFrame(
    {'answer': answers,
     'byline': bylines,
     'date': dates,
     'url': urls,
     }
)
```
